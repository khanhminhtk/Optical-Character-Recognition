# extract_data

```python
def extract_data(file):
  images_labels = []
  images_sizes = []
  boxes = []
  texts = []
  tree = ET.parse(file)
  root = tree.getroot()

  for image in root:
    images_labels.append(image[0].text)
    images_sizes.append([float(image[1].attrib['x']), float(image[1].attrib['y'])])
    for i in image.findall('taggedRectangles'):
      boxes.append([])
      texts.append([])
      for box in i:
        if not box[0].text.isalnum():
          continue
        if 'é' in box[0].text.lower() or 'ñ' in box[0].text.lower():
          continue
        texts[-1].append(box[0].text.lower())
        boxes[-1].append([float(box.attrib['x']), float(box.attrib['y']), float(box.attrib['width']), float(box.attrib['height'])])
  return images_labels, images_sizes, boxes, texts

```

```python
images_labels, images_sizes, boxes, texts = extract_data(data_raw_file)
```

# transform data for yolov8


```python
def transform_yolo_format(images_sizes, boxes, images_labels):
  yolo_data = []

  for size, box, lable in zip(images_sizes, boxes, images_labels):
    image_with, image_height = size
    yolo_labels = []
    for i in range(0, len(box)):
      x, y, w, h = box[i]
      x_center = (x + w / 2) / image_with
      y_center = (y + h / 2) / image_height
      w = w / image_with
      h = h / image_height

      class_id = 0

      yolo_labels.append(f'{class_id} {x_center} {y_center} {w} {h}')
    yolo_data.append((lable, yolo_labels))
  return yolo_data
```

```python
yolo_data = transform_yolo_format(images_sizes, boxes, images_labels)
yolo_data
```

```python
from sklearn.model_selection import train_test_split
seed = 0
val_size = 0.2
test_size = 0.125
is_shuffle = True

train_data, test_data = train_test_split(
    yolo_data,
    test_size=val_size,
    random_state=seed,
    shuffle=is_shuffle
)
test_data, val_data = train_test_split(
    test_data,
    test_size=test_size,
    random_state=seed,
    shuffle=is_shuffle
)
```

```python
import shutil
def load_data(data, src_img_dir, save_dir):
    os.makedirs(save_dir, exist_ok=True)

    os.makedirs(os.path.join(save_dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(save_dir, 'labels'), exist_ok=True)

    for image_path, yolov8_labels in data:
        shutil.copy(
            os.path.join(src_img_dir, image_path),
            os.path.join(save_dir, 'images')
        )

        image_name = os.path.basename(image_path)
        image_name = os.path.splitext(image_name)[0]

        with open(os.path.join(save_dir, 'labels', f"{image_name}.txt"), 'w') as f:
            for label in yolov8_labels:
                f.write(f"{label}\n")
```

```python
folder = "/content/drive/MyDrive/Code/cv/OCR/data/yolo_data"
os.makedirs(folder, exist_ok = True)
save_train_dir = os.path.join(folder, 'train')
save_val_dir = os.path.join(folder, 'val')
save_test_dir = os.path.join(folder, 'test')

data_raw_file = '/content/drive/MyDrive/Code/cv/OCR/data_raw/SceneTrialTrain'
load_data(train_data, data_raw_file, save_train_dir)
load_data(test_data, data_raw_file, save_test_dir)
load_data(val_data, data_raw_file, save_val_dir)
```

```python
import cv2
import json
import matplotlib.pyplot as plt

def visualize_bbox(image_labels, predictions, conf_thres=0.8, font=cv2.FONT_HERSHEY_SIMPLEX):
    img = cv2.imread(image_labels)
    h, w = img.shape[:2]

    for prediction in predictions:
        conf_score = prediction['confidence']

        if conf_score < conf_thres:
            continue

        bbox = prediction['box']
        xmin = int(bbox['x1'])
        ymin = int(bbox['y1'])
        xmax = int(bbox['x2'])
        ymax = int(bbox['y2'])

        cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (0, 255, 0), 3)


        text = f"{conf_score:.2f}"
        (text_width, text_height), _ = cv2.getTextSize(text, font, 1, 2)

        cv2.rectangle(img, (xmin, ymin - text_height - 5), (xmin + text_width, ymin), (0, 255, 0), -1)
        cv2.putText(img, text, (xmin, ymin - 5), font, 1, (0, 0, 0), 2)


    return img
```

```python

img_path = "/content/drive/MyDrive/Code/cv/OCR/data/yolo_data/test/images/IMG_1282.JPG"

results = model(img_path, verbose=False)
predictions = json.loads(results[0].to_json())
print(predictions[0])
visualized_img = visualize_bbox(img_path, predictions, 0.75)

%matplotlib inline
plt.imshow(visualized_img)
plt.axis('off')
plt.show()
```
